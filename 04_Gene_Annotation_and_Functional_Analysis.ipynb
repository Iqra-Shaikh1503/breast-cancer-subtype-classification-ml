```python
import gzip
import pandas as pd

# Read the annotation file
with gzip.open(
    r"C:\Research_Paper_Project\GPL570.annot.gz",
    "rt",
    encoding="utf-8",
    errors="ignore"
) as f:

    lines = f.readlines()

# Find where the platform table begins
start = None
for i, line in enumerate(lines):
    if line.startswith("!platform_table_begin"):
        start = i + 1
        break

# Load only the platform table
from io import StringIO

annotation = pd.read_csv(
    StringIO("".join(lines[start:])),
    sep="\t"
)

print(annotation.head())
print(annotation.columns)
```

```python
annotation = annotation[
    ["ID", "Gene symbol", "Gene title"]
]

annotation.head()
```

```python
top_genes = pd.read_csv(
    r"C:\Research_Paper_Project\Project\results\top_ranked_genes.csv"
)

annotated = top_genes.merge(
    annotation,
    left_on="Probe_ID",
    right_on="ID",
    how="left"
)

annotated.head(20)
```

```python
annotated.to_csv(
    r"C:\Research_Paper_Project\Project\results\top_genes_annotated.csv",
    index=False
)

print("Annotated genes saved successfully!")
```
